```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Create synthetic data points
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_redundant=0, n_clusters_per_class=1)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel='linear')
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

### RBF Kernel

```python
X,y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_redundant=0, n_clusters_per_class=2)
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
rbf=SVC(kernel='rbf')
rbf.fit(X_train, y_train)
```

```python
y_pred = rbf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
### Polynomial SVC
```

```python
polynomial=SVC(kernel='poly')
polynomial.fit(X_train, y_train)
y_pred = polynomial.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
### Sigmoid SVC

sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(X_train, y_train)
y_pred = sigmoid.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'C': [0.1, 1, 10, 100, 100],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf']}
    
grid = GridSearchCV(SVC(), param_grid=param_grid, cv=5, verbose=3, refit=True)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
